```python
import numpy as np
import pandas as pd
df=pd.read_csv("bitcoin.csv")
df.head()
```

```python
df
```

```python
#removing date column in this
#1: Indicates the axis along which the drop operation will be performed. Here, '1' represents the column axis.
#inplace=True: Modifies the DataFrame 'df' in-place, meaning the changes will be applied to the DataFrame itself, rather than creating a new DataFrame. If you set inplace=False (default value), a new DataFrame with the column dropped will be returned, and the original DataFrame 'df' will remain unchanged.
df.drop(['Date'],1, inplace=True)
```

```python
predictionDays=30
#create another colum shifted 'n' units up
df['Prediction']=df[['Price']].shift(-predictionDays)
df.head()
```

```python
df.tail()
```

```python
#here we will convery the data frame into a numpy array and drp the prediction column
x=np.array(df.drop(['Prediction'],1))
#remove the last 'n' rows where 'n' is the predictionDays
x=x[:len(df)-predictionDays]
print(x)
```

```python
# Create the dependent data set
# convert the data frame into a numpy array
y = np.array(df['Prediction'])
# Get all the values except last 'n' rows
y = y[:-predictionDays]
print(y)
```

```python
x.shape
```

```python
y.shape
```

```python
# Split the data into 80% training and 20% testing
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(x,y, test_size = 0.2)
# set the predictionDays array equal to last 30 rows from the original data set
predictionDays_array = np.array(df.drop(['Prediction'],1))[-predictionDays:]
print(predictionDays_array)
```

```python
from sklearn.svm import SVR
svr_rbf=SVR(kernel='rbf', C=1e3, gamma=0.00001)
svr_rbf.fit(xtrain,ytrain)
```

```python
svr_rbf_confidence=svr_rbf.score(xtest,ytest)
print('SVR_RBF accuracy:',svr_rbf_confidence)
```

```python
svm_prediction=svr_rbf.predict(xtest)
print(svm_prediction)
print(ytest)
```

```python
#model prediction for next 30 days
svm_prediction = svr_rbf.predict(predictionDays_array)
print(svm_prediction)
print()
#Print the actual price for bitcoin for last 30 days
print(df.tail(predictionDays))
```
